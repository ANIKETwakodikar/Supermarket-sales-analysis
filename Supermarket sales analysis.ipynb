```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/home/aniket/Desktop/supermarket_sales - Sheet1.csv')
df
```

# Payment modes breakdown

```python
l=list(df.Payment)
plt.subplot(2,2,1)
plt.pie([(l.count('Ewallet')/len(l)*100),(l.count('Cash')/len(l)*100),(l.count('Credit card')/len(l)*100)], labels=['Ewallet', 'Cash', 'Credit card'],autopct='%1.1f%%')
plt.title('All 3 branches combined.')

#Branch A
plt.subplot(2,2,2)
plt.pie(list(df.loc[(df['Branch']=='A')].Payment.value_counts()),labels=df.Payment.unique(),autopct='%1.1f%%')
plt.title('Branch A ')

#Branch B
plt.subplot(2,2,3)
plt.pie(list(df.loc[(df['Branch']=='B')].Payment.value_counts()),labels=df.Payment.unique(),autopct='%1.1f%%')
plt.title('Branch B ')

#Branch C
plt.subplot(2,2,4)
plt.pie(list(df.loc[(df['Branch']=='C')].Payment.value_counts()),labels=df.Payment.unique(),autopct='%1.1f%%')
plt.title('Branch  ')


```

# Male vs Female customer ratio

```python
plt.subplot(2,2,1)
plt.bar(df.Gender.unique(),df.Gender.value_counts())
plt.title('All 3 branches')

plt.subplot(2,2,2)
plt.bar(df.loc[(df['Branch']=='A')].Gender.unique(),[list(df.loc[(df['Branch']=='A')].Gender).count('Female'),list(df.loc[(df['Branch']=='A')].Gender).count('Male')])
plt.title('Branch A')

plt.subplot(2,2,3)
plt.bar(df.Gender.unique(),[list(df.loc[(df['Branch']=='B')].Gender).count('Female'),list(df.loc[(df['Branch']=='B')].Gender).count('Male')])
plt.title('Branch B')

plt.subplot(2,2,4)
plt.bar(df.Gender.unique(),[list(df.loc[(df['Branch']=='C')].Gender).count('Female'),list(df.loc[(df['Branch']=='C')].Gender).count('Male')])
plt.title('Branch C')
```

# Male vs Female sales profit contribution

```python
plt.subplot(2,2,1)
plt.bar(['Male','Female'],[sum(df.loc[df['Gender']=='Male']['gross income']),sum(df.loc[df['Gender']=='Female']['gross income'])])
plt.title('All 3 branches')

plt.subplot(2,2,2)
plt.bar(['Male','Female'],[sum(df.loc[(df['Gender']=='Male') & (df['Branch']=='A')]['gross income']),sum(df.loc[(df['Gender']=='Female') & (df['Branch']=='A')]['gross income'])])
plt.title('Branch A')

plt.subplot(2,2,3)
plt.bar(['Male','Female'],[sum(df.loc[(df['Gender']=='Male') & (df['Branch']=='B')]['gross income']),sum(df.loc[(df['Gender']=='Female') & (df['Branch']=='B')]['gross income'])])
plt.title('Branch B')

plt.subplot(2,2,4)
plt.bar(['Male','Female'],[sum(df.loc[(df['Gender']=='Male') & (df['Branch']=='C')]['gross income']),sum(df.loc[(df['Gender']=='Female') & (df['Branch']=='C')]['gross income'])])
plt.title('Branch C')


```

# Branch wise profit comparison

```python
plt.bar(['A (Yangon)','B (Mandalay)','C (Naypyitaw)'],[sum(df.loc[(df['Branch']=='A')]['gross income']),sum(df.loc[(df['Branch']=='B')]['gross income']),sum(df.loc[(df['Branch']=='C')]['gross income'])])
```

# Member vs normal customer profit contribution

```python
plt.subplot(2,2,1)
plt.pie([sum(df.loc[df['Customer type']=='Member']['gross income']),sum(df.loc[df['Customer type']=='Normal']['gross income'])],labels=['Member','Customer'],autopct='%1.1f%%')
plt.title('All three branches')

plt.subplot(2,2,2)
plt.pie([sum(df.loc[(df['Customer type']=='Member') & (df['Branch']=='A') ]['gross income']),sum(df.loc[(df['Customer type']=='Normal') & (df['Branch']=='A') ]['gross income'])],labels=['Member','Customer'],autopct='%1.1f%%')
plt.title('Branch A')

plt.subplot(2,2,3)
plt.pie([sum(df.loc[(df['Customer type']=='Member') & (df['Branch']=='B')]['gross income']),sum(df.loc[(df['Customer type']=='Normal') & (df['Branch']=='B') ]['gross income'])],labels=['Member','Customer'],autopct='%1.1f%%')
plt.title('Branch c')

plt.subplot(2,2,4)
plt.pie([sum(df.loc[(df['Customer type']=='Member') & (df['Branch']=='C') ]['gross income']),sum(df.loc[(df['Customer type']=='Normal') & (df['Branch']=='C') ]['gross income'])],labels=['Member','Customer'],autopct='%1.1f%%')
plt.title('Branch C')

```

# Average rating of each product line type`

```python
t=[]
for i in df['Product line'].unique():
       t.append(df.loc[df['Product line']==i]['Rating'].mean())
plt.subplot(4,1,1)
plt.barh(df['Product line'].unique(),t)
plt.title('All 3 branches')

A=[]
for i in df['Product line'].unique():
       A.append(df.loc[(df['Product line']==i) & (df['Branch']=='A')]['Rating'].mean())
plt.subplot(4,1,2)
plt.barh(df['Product line'].unique(),A)
plt.title('Branch A')

B=[]
for i in df['Product line'].unique():
       B.append(df.loc[(df['Product line']==i) & (df['Branch']=='B')]['Rating'].mean())
plt.subplot(4,1,3)
plt.barh(df['Product line'].unique(),B)
plt.title('Branch B')

C=[]
for i in df['Product line'].unique():
       C.append(df.loc[(df['Product line']==i) & (df['Branch']=='C')]['Rating'].mean())
plt.subplot(4,1,4)
plt.barh(df['Product line'].unique(),B)
plt.title('Branch C')
    
```
